```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

## Read in the data from the NASDAQ API

The most recent stock data from the API is dated to March 27, 2018. So we're adding a start_date of 2013-03-27 to get 5 years' worth of data.

```python
VZ_url = "https://data.nasdaq.com/api/v3/datasets/WIKI/VZ.csv?start_date=2013-03-27"
VZ_df = pd.read_csv(VZ_url)

VZ_df
```

```python
ATT_url = "https://data.nasdaq.com/api/v3/datasets/WIKI/T.csv?start_date=2013-03-27"
ATT_df = pd.read_csv(ATT_url)

ATT_df
```

```python
TMUS_url = "https://data.nasdaq.com/api/v3/datasets/WIKI/TMUS.csv?start_date=2013-03-27"
TMUS_df = pd.read_csv(TMUS_url)

TMUS_df
```

```python
# Check the data types for one of the dataframes (since they all come from the same API)
VZ_df.info()
```

```python
# Change the 'Date' data in each dataframe to datetime/timestamp, instead of object
VZ_df["Date"] = pd.to_datetime(VZ_df["Date"], format="%Y/%m/%d")
ATT_df["Date"] = pd.to_datetime(ATT_df["Date"], format="%Y/%m/%d")
TMUS_df["Date"] = pd.to_datetime(TMUS_df["Date"], format="%Y/%m/%d")

# Double check that the data type change went through:
VZ_df.info()
```

```python
# Export the dataframes as CSV files for future reference
VZ_df.to_csv("Resources/VZ.csv")
ATT_df.to_csv("Resources/ATT.csv")
TMUS_df.to_csv("Resources/TMUS.csv")
```

## Closing Price Trends and Comparison

### Closing Prices: Verizon


5-Year Period

```python
# Graph the past 5 years' closing prices
y_5yr = VZ_df["Close"]
x_5yr = VZ_df["Date"]

# initial_close needs to be recalculated
initial_close = y_5yr.iloc[-1]
recent_close = y_5yr.iloc[0]

plt.figure(figsize=(24, 8))
plt.title("Verizon (VZ) Closing Prices, 2013-03-27 to 2018-03-27", fontsize=20)
plt.plot(x_5yr, y_5yr, color='red')
plt.savefig("Images/VZ_closing_5yr.jpg")

print(VZ_df.describe()["Close"])
print(f"Initial Closing Price: {initial_close}, ({x_5yr.iloc[-1]})")
print(f"Latest Closing Price: {recent_close}, ({x_5yr.iloc[0]})")
print(f"Total Change in Price: {round(recent_close - initial_close, 2)}")
```

```python
# Graph just the past year's worth of data
mask = (VZ_df["Date"] > '2017-03-26') & (VZ_df["Date"] <= '2018-03-27')
VZ_1yr_df = VZ_df.loc[mask]

y_1yr = VZ_1yr_df["Close"]
x_1yr = VZ_1yr_df["Date"]

# initial_close needs to be recalculated
initial_close = y_1yr.iloc[-1]

plt.figure(figsize=(24, 8))
plt.title("Verizon (VZ) Closing Prices, 2017-03-27 to 2018-03-27", fontsize=20)
plt.plot(x_1yr, y_1yr, color='red')
plt.savefig("Images/VZ_closing_1yr.jpg")

print(VZ_1yr_df.describe()["Close"])
print(f"Initial Closing Price: {initial_close}, ({x_1yr.iloc[-1]})")
print(f"Latest Closing Price: {recent_close}, ({x_5yr.iloc[0]})")
print(f"Total Change in Price: {round(recent_close - initial_close, 2)}")
```

```python
# Graph the past month's data (from this dataset)
mask = (VZ_df["Date"] > '2018-02-26') & (VZ_df["Date"] <= '2018-03-27')
VZ_1mn_df = VZ_df.loc[mask]

y_1mn = VZ_1mn_df["Close"]
x_1mn = VZ_1mn_df["Date"]

# initial_close needs to be recalculated
initial_close = y_1mn.iloc[-1]

plt.figure(figsize=(16, 8))
plt.title("Verizon (VZ) Closing Prices, 2018-02-27 to 2018-03-27", fontsize=20)
plt.plot(x_1mn, y_1mn, color='red')
plt.savefig("Images/VZ_closing_1mn.jpg")

print(VZ_1mn_df.describe()["Close"])
print(f"Initial Closing Price: {initial_close}, ({x_1mn.iloc[-1]})")
print(f"Latest Closing Price: {recent_close}, ({x_5yr.iloc[0]})")
print(f"Total Change in Price: {round(recent_close - initial_close, 2)}")
```

### Comparing Closing Prices of All Stocks

#### Past five years

```python
# Getting all three compared
VZ = VZ_df["Close"]
ATT = ATT_df["Close"]
TMUS = TMUS_df["Close"]
x = VZ_df["Date"]

plt.figure(figsize=(20, 10))
plt.plot(x, VZ, color='red', label='Verizon (VZ)')
plt.plot(x, ATT, color='steelblue', label='AT&T (T)')
plt.plot(x, TMUS, color='fuchsia', label='T-Mobile (TMUS)')
plt.title("All Stocks' NASDAQ Closing Prices, 2013-2018", fontsize=20)
plt.xlabel("Date", fontsize=14)
plt.ylabel("Closing Price", fontsize=14)
plt.legend(loc="best")
plt.savefig("Images/all_closing_5yr.jpg")
```

Comparing all three stocks across the most recent five years in the dataset, Verizon and AT&T's closing prices have remained largely stagnant, with their respective changes in prices being -1.63 and -1.70, while T-Mobile has experienced significant growth. At some point in late 2016, T-Mobile's stock price overtook Verizon's, which was the stock price leader until that point.

```python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

### Read in the data from the NASDAQ API

A start date of 2008-03-27 was added to the API call in order to receive 10 years' worth of stock data, since the most recent stock data retrieved is dated to 2018-03-27.

```python
VZ_url = "https://data.nasdaq.com/api/v3/datasets/WIKI/VZ.csv?start_date=2008-03-27"
VZ_df = pd.read_csv(VZ_url)

VZ_df
```

```python
ATT_url = "https://data.nasdaq.com/api/v3/datasets/WIKI/T.csv?start_date=2008-03-27"
ATT_df = pd.read_csv(ATT_url)

ATT_df
```

```python
TMUS_url = "https://data.nasdaq.com/api/v3/datasets/WIKI/TMUS.csv?start_date=2008-03-27"
TMUS_df = pd.read_csv(TMUS_url)

TMUS_df
```

```python
# Check the data types for one of the dataframes (since they all come from the same API)
VZ_df.info()
```

```python
# Change the 'Date' data in each dataframe to datetime/timestamp, instead of object
VZ_df["Date"] = pd.to_datetime(VZ_df["Date"], format="%Y/%m/%d")
ATT_df["Date"] = pd.to_datetime(ATT_df["Date"], format="%Y/%m/%d")
TMUS_df["Date"] = pd.to_datetime(TMUS_df["Date"], format="%Y/%m/%d")
```

```python
# Export the dataframes as CSV files for future reference
VZ_df.to_csv("Resources/VZ.csv")
ATT_df.to_csv("Resources/ATT.csv")
TMUS_df.to_csv("Resources/TMUS.csv")
```

### Comparison of Closing Prices

```python
# Line plot for Verizon's closing prices over time
y = VZ_df["Close"]
x = VZ_df["Date"]

plt.figure(figsize=(24,12))
plt.title("Verizon (VZ) Closing Prices, 2008-03-27 to 2018-03-27")
plt.plot(x, y)

print(VZ_df.describe()["Close"])
```

```python
# Graph just the past year's worth of data
mask = (VZ_df["Date"] > '2017-03-26') & (VZ_df["Date"] <= '2018-03-27')
VZ_1year_df = VZ_df.loc[mask]

y = VZ_1year_df["Close"]
x = VZ_1year_df["Date"]

plt.figure(figsize=(24,12))
plt.title("Verizon (VZ) Closing Prices, 2017-03-27 to 2018-03-27")
plt.plot(x, y)

print(VZ_1year_df.describe()["Close"])
```

```python
# Graph the past month's data (from this dataset)
mask = (VZ_df["Date"] > '2018-02-26') & (VZ_df["Date"] <= '2018-03-27')
VZ_1week_df = VZ_df.loc[mask]

y = VZ_1week_df["Close"]
x = VZ_1week_df["Date"]

plt.figure(figsize=(24,12))
plt.title("Verizon (VZ) Closing Prices, 2018-02-27 to 2018-03-27")
plt.plot(x, y)

print(VZ_1week_df.describe()["Close"])
```
